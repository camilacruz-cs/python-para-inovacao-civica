### Habilitando filtros que serão usados na análise de dados.

```python
#Armazenando o primeiro filtro que será usado
filtro_cargo = data_completo['DS_CARGO'] == 'GOVERNADOR'
```

### Qual a faixa etária média das pessoas candidatas?

```python
df_raca_genero['NR_IDADE_DATA_POSSE'].unique
```

```python
#Aqui estamos estabelecendo idades aleatóriamente, pois não temos uma coluna específica com essa informação.
cortes = [18, 25, 45, 55, 65, 100]
```

```python
#Aqui nós vamos estabelecer faixas para os cortes

categorias = ['18 a 24 anos', '25 a 34 anos', '35 a 44 anos', '55 a 64 anos', 'acima de 64 anos']
```

## Panda.cut()
### Use cut when you need to segment and sort data values into bins. This function is also useful for going from a continuous variable to a categorical variable. For example, cut could convert ages to groups of age ranges.

```python
#Label são as etiquetas que serão os cortes
df_raca_genero['FAIXA_ETARIA'] = pd.cut(df_raca_genero['NR_IDADE_DATA_POSSE'],cortes, labels=categorias)
```

```python
#Agora a FAIXA ETARIA está contida no DataFrame que foi criado no início.
df_raca_genero.sample(3)
```

```python
#Verificando se a categoria deu certo

df_raca_genero[df_raca_genero['NR_IDADE_DATA_POSSE']== 55]
```

### Qual é a faixa etária mais predominante?

```python
df_raca_genero['FAIXA_ETARIA'].mode()
```

### Visualize graficamente a distribuição da variável DS_COR_RACA em 2022

```python
df_raca = df_governadores(['CD_COR_RACA', 'DS_COR_RACA']).copy()
```

```python
df_raca[['CD_COR_RACA', 'DS_COR_RACA']].groupby(by=['DS_COR_RACA']).count()
```

```python
frequencia = df_raca[['DS_COR_RACA']].value_counts(sort=False).tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar('DS_COR_RACA', frequencia)
plt.xticks(rotation=45)
plt.title('Distribuição por Raça em 2022')
plt.ylabel('Quantidade de candidatura')

plt.show()
```
