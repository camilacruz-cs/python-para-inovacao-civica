```python
#Aqui eu seleciono todas as colunas que eu quero que sejam associadas ao meu filtro desejado.
colunas_desejadas = ['SG_PARTIDO', 'DS_GENERO', 'DS_COR_RACA']
filtro = dados_completos['SG_PARTIDO'] == 'AP'
```

```python
candidatos_sexo_raca = dados_completos.loc[filtro, colunas_desejadas]
candidatos_sexo_raca
```

```python
#Agrupamento pela variÃ¡vel gÃªnero.
candidatos_sexo_raca.groupby(['DS_GENERO']).count()
```

```python
candidatos_sexo_raca.groupby(['DS_COR_RACA']).count()
```

```python
candidatos_sexo_raca[candidatos_sexo_raca['DS_GENERO'] == 'FEMININO'].groupby(
    ['DS_GENERO', 'DS_COR_RACA']).count()
```

```python
#E se quisermos descobrir, quais os partidos que possuem o maior nÃºmero de representaÃ§Ã£o em cada grupo social?

candidatos_sexo_raca[candidatos_sexo_raca['DS_GENERO'] == 'FEMININO'].groupby(
    ['DS_GENERO', 'DS_COR_RACA']).max()
```
