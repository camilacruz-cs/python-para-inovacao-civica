### Listas

* São uma sequência de variáveis que queremos guardar ou reservar.

* Toda contagem em listas começa no ZERO. Logo, [0, 1, 2...]

```python
frutas = ['pera', 'uva', 'maça', 'laranja']
```

```python
#Pegar o segundo elemento.
#Idexaçao no Python inicia no ZERO [0]

frutas[2]
```

```python
#Pegando o último elemento

frutas[3]
```

### Fatiamento de listas

* Pegando apenas trechos da lista.

```python
frutas[0:2]
```

```python
#Pegando a primeira
frutas[:1]
```

```python
#Tamanho de listas
len(frutas)
```

###Dicionário

* É uma lista mais 'avançada.
* Chave: par-valor

```python
salada_de_frutas = {'ingredientes': frutas,
                    'preparo':'corte tudo e misture',
                    'tempo':60
                    }
```

```python
#pegar um item
salada_de_frutas['ingredientes']
```

```python
#adicionar item
salada_de_frutas['porções'] = 7
```

```python
#excluir item
del salada_de_frutas['tempo']
```

```python
print(salada_de_frutas)
```

## Estrutura de repetição `For`

* É um looping.

```python
#Iterando em listas

for frutas in frutas:
  print(frutas)
```

```python
#Pegar as chaves do dicionário

for chave in salada_de_frutas:
  print(salada_de_frutas[chave])
```

```python
#Iterando dentro de um texto

for letra in evento:
  print(letra)
```

## Estrutura Condicional `if`

* Traz se é verdadeiro ou falso seu questionamento.

```python
if 4 > 3:
  print('É maior sim')
```

```python
if 4 > 7:
  print('É maior sim')
else:
  print('claro que nao é')
```

```python
idade = 21

if idade > 30:
  print('Sou adulto')
elif idade > 19 & idade <= 30:
  print('Sou jovem')
else:
  print('Sou adolescente')
```

## Agrupar e agregar dados


```python
df_empresas.groupby('Ano')
```

```python
df_empresas.groupby('Ano').sum()

```

```python
df_empresas[['Ano','Nº Trab. cond. analog escravo']].groupby('Ano').sum().sort_values('Nº Trab. cond. analog escravo', ascending=False)
```

## Bora Investigar? 🔎🕵🏾‍♀️

```python
#Total de pessoas  libertas do trabalho escravo de 2010 a 2020
df_empresas['Nº Trab. cond. analog escravo'].sum()
```

```python
#rancking de empresas (CNAE)
df_empresas['Descrição CNAE'].value_counts().head(5)
```

```python
#Qual ano que mais houveram pessoas libertadas

df_empresas[['Ano','Nº Trab. cond. analog escravo']].groupby('Ano').sum().sort_values('Nº Trab. cond. analog escravo', ascending=False)
```
