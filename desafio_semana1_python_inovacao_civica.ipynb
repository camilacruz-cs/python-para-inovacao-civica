```python
print(consulta_cand)
```

# Crie um laço (loop) para ler todos os arquivos da lista (usando a biblioteca Pandas) e armazenar os dataframes em uma lista

```python
from IPython.utils import encoding
lista_estados = []

for dados_consulta in consulta_cand:
  df_lista = pd.read_csv(dados_consulta, sep=';', encoding='latin_1')
  lista_estados.append(df_lista)
```

```python
print(lista_estados)
```

# Concatene todos os dataframes da lista

```python
df_completo = pd.concat(lista_estados, axis=0, ignore_index=True)
```
